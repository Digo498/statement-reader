```python
import pdfplumber

# Box format in (x_top, y_top, x_bot, y_bot)
regular_page_box = (70, 210, 400, 730)
initial_page_box = (70, 440, 400, 730)

text = ''
# Open the PDF file
with pdfplumber.open('statement.pdf') as pdf:
    for page in pdf.pages:
        if page.page_number == 1:
            rect = initial_page_box
        else:
            rect = regular_page_box

        # Extract text only from the specified rectangle
        text += page.crop(rect).extract_text(x_tolerance=2)

# Process the extracted text
lines = text.split('\n')
print(lines)
```

```python
transactions = []

pattern = r'(\w{3}\s+\d{1,2})\s+(.+) (\d+\.\d+)'

for i, line in enumerate(lines):
    match = re.match(pattern, line)
    print(line)
    if match:
        date = match.group(1)
        transaction_type = match.group(2)
        amount = match.group(3)

        # Create a dictionary for the transaction and add it to the list
        transaction = {
            'Date': date,
            'Transaction Type': transaction_type,
            'Amount': amount,
            'Name': lines[i+1],
        }
        transactions.append(transaction)


transactions
```
